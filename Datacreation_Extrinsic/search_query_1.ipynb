**`Web scrapping with the query`**

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup as bs 
import pandas as pd
```

```python
# dataframe from Dataset_creation(kmeans+GoogleApi).ipynb
data=pd.read_csv('All_query_aggregatev1.csv')
data.head(5)
```

```python
query_urls=list(data['url'])
print(len(query_urls))
```

```python
#formatting the url's and removing the white spaces
#considering the urls only with https
main_text=[]
for query in query_urls:
  list1=[]
  text=""
  for  i in query.split(","):
    new=(i.replace("'| ' ","").replace("[","").replace("]","").replace(" ' ","").strip())
    new=new[1:len(new)-1]
    if new[:5]=="https":
      text=str(new)
      list1.extend([text])
  #list1=list1[:40]
  main_text.append(list1)
```

```python
data['new_url']=main_text
data.head(3)
```

```python
#count the number of urls per query
data['new_length']=data['new_url'].apply(lambda x: len(x))

count=0
cnt_list=[]
for i in range(100):
  val=len(data['new_url'][i])
  count=count + val
  cnt_list.append(count)
print(cnt_list)

data.head(3)
```
