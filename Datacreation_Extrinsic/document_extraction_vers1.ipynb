```python
#Finding the top 10 similar sentences based on the anti-sterotypical sentence 

from sklearn.feature_extraction.text import CountVectorizer

cluster_doc_list=[]

for i,j in cluster_info.iterrows():
    temp_list=[]
    anti_stereo_text=j['anti_stereo_sentence']
    print(i)
    sample_doc=re.sub("[^a-zA-Z.]", " ", str(j['document']))
    sample_lines=sample_doc.split(".")
    vec = CountVectorizer()
    #print("in")
    
    text_matrix = vec.fit_transform(sample_lines)
    anti_stereo_text_vec=vec.transform([anti_stereo_text])
    
    list1=[]
    
    for k in range(text_matrix.shape[0]):
        result = cosine_similarity(anti_stereo_text_vec,text_matrix[k])
        list1.append(result)
    
    list1=np.array(list1)
    
    list1=list1.flatten()
    indices = np.argsort(list1)[::-1]
    
    for i in indices[:10]:
        temp_list.append(sample_lines[i])
        
    cluster_doc_list.append(temp_list)
    
print("End Run---------------------------------------------------")
print("Length of cluster list : ",len(cluster_doc_list))

cluster_info['top10sentence']=cluster_doc_list

cluster_info.to_csv('Final_dataframe18Nov_vers2.csv',index=False)

cluster_info.head(5)


    
    
    
        

    
    

    
```
